```python
data=pd.read_csv('/kaggle/input/dataset/dynamic_pricing.csv')
```

```python
# Define demand cutoff and calculate demand metric

demand_cutoff = 1.15
data['Demand_Metric'] = data['Number_of_Riders'] / data['Number_of_Drivers']

# Calculate revenue when demand > supply
data['Revenue'] = data.apply(lambda row: row['Number_of_Riders'] * row['Historical_Cost_of_Ride'] if row['Demand_Metric'] > 1 else 0, axis=1)

# Adjust cost of ride based on the demand metric using the formula: 
# Adjusted Cost = Historical Cost * (1 + ((Demand Metric - Demand Cutoff) / Demand Cutoff))
data['Adjusted_Ride_Cost'] = data.apply(lambda row: row['Historical_Cost_of_Ride'] * (1 + ((row['Demand_Metric'] - demand_cutoff) / demand_cutoff)) if row['Demand_Metric'] > demand_cutoff else row['Historical_Cost_of_Ride'], axis=1)

# Display the relevant columns for analysis
result = data[['Number_of_Riders', 'Number_of_Drivers', 'Demand_Metric', 'Historical_Cost_of_Ride', 'Revenue', 'Adjusted_Ride_Cost']]
result.head()
```

The revenue can be calculated using the formula:
**Revenue=Number of Riders×Historical Cost of Ride**

**Revenue Calculation:**

Revenue was calculated for every instance where the demand metric (number of riders divided by number of drivers) is greater than 1, indicating that demand exceeds supply.

**Adjusted Ride Cost:**

When the demand metric exceeds the demand cutoff (**1.15** in this case), the historical ride cost was adjusted upwards. If the demand metric was below the cutoff, the original ride cost was kept.

Adjusted Cost = Historical Cost * (1 + ((Demand Metric - Demand Cutoff) / Demand Cutoff))
